```python
### 필요한 라이브러리 / 함수 임폴트
# import PIL
# from PIL.Image import open
# import PIL.Image as Image
from PIL import Image # 공식 문서 상의 표현으로 주로 사용
import matplotlib.pyplot as plt
import numpy as np
```

## 이미지 읽기 및 출력

```python
# 이미지 파일 경로 설정
file_path = '/content/drive/MyDrive/딥러닝/Lenna.png'

# image.open() 함수 이용 → 이미지 읽기
img = Image.open(file_path)
print(f'Image.open 함수를 사용한 결과값 : {img}')

print('-'*80)

# numpy 배열로 변환
img_array = np.array(img)
print(f'numpy 배열로 변환한 결과 : \n{img_array}') # 픽셀값 정보

print('-'*80)

# 이미지 출력하기
plt.imshow(img)
plt.axis('off') # 축의 값 제거
plt.show()
```

## 이미지의 속성(정보) 출력

```python
# 이미지 파일 경로 설정
file_path = '/content/drive/MyDrive/딥러닝/Lenna.png'

# image.open() 함수 이용 → 이미지 읽기
img = Image.open(file_path)
print(f'Image.open 함수를 사용한 결과값 : {img}')
# 결과값 해석이 어려우므로 속성을 별도로 출력해서 확인해야함

print('-'*80)

# 이미지의 속성 확인하기
print(f'파일 경로(이름) : {img.filename}')
print('-'*80)
print(f'파일 형식 : {img.format}')
print('-'*80)
print(f'색상 모드 : {img.mode}') # 흑백(GRAY) or 컬러(RGB) or 투명도(RGBA)
print('-'*80)
print(f'이미지의 크기 : {img.size}') # 전체 크기 (해상도)
print('-'*80)
print(f'이미지의 너비 : {img.width}')
print('-'*80)
print(f'이미지의 높이 : {img.height}')
print('-'*80)
print(f'이미지의 자료형 : {type(img)}')
```

## 이미지의 크기 조절

```python
# 이미지 파일 경로 설정
file_path = '/content/drive/MyDrive/딥러닝/Lenna.png'

# image.open() 함수 이용 → 이미지 읽기
img = Image.open(file_path)
print(f'Image.open 함수를 사용한 결과값 : {img}')

print('-'*80)

# 이미지의 크기 확인
print(f'이미지의 크기(해상도) 확인 : {img.size}')

print('-'*80)

# 이미지의 크기(해상도) 조절 → resize((width, height))
# 보통은 넘파이 배열로 바꿔서 숫자값을 조절하나 PIL은 넘파이 배열이 아니어도 가능 - resize((튜플))
# 핵심 : 픽셀 크기가 고정된 것이 아님 1pixel의 크기가 다를 수 있음. 큰 픽셀은 큰 물체 밖에 담을 수 없음
resized_img = img.resize((256,256))

# 이미지 출력
print('원본 이미지')
plt.imshow(img)
plt.show()
print('-'*80)
print('크기를 1/4로 줄인 이미지')
plt.imshow(resized_img)
plt.show()
```
