# BertWordPiece Tokenizer

## 단어 사전 생성

```python
### 필요한 라이브러리 설치
!pip install tokenizers
```

```python
### 필요한 함수 임폴트
from tokenizers import BertWordPieceTokenizer
```

```python
### Tokenizer 모델 생성
tokenizer1 = BertWordPieceTokenizer()
tokenizer2 = BertWordPieceTokenizer()
```

```python
### 단어 사전 생성

# 파일 경로 설정
file_path = '/content/drive/MyDrive/딥러닝/nsmc.txt'

# train() 함수 호출 → 단어 사전 생성
tokenizer1.train(
    files = file_path,
    vocab_size = 5000,
)

tokenizer2.train(
    files = file_path,
    vocab_size = 20000,
)
```

```python
### 생성된 단어 사전 불러오기

# 단어 사전의 크기 : 5000개
vocab1 = tokenizer1.get_vocab()

# 단어 사전의 크기 : 20000
vocab2 = tokenizer2.get_vocab()

# 결과 확인하기
print(f'단어 사전 확인 : \n{vocab1}')
print('-'*80)
print(f'단어 사전 확인 : \n{vocab2}')
```

```python
### 단어 사전 정렬 --> value 기준 --> 오름차순 정렬
sorted_vocab1 = sorted(vocab1.items(), key=lambda x:x[1])
print(f'오름차순으로 정렬된 단어 사전 : \n{sorted_vocab1}')

print('-'*80)

sorted_vocab2 = sorted(vocab2.items(), key=lambda x:x[1])
print(f'오름차순으로 정렬된 단어 사전 : \n{sorted_vocab2}')
```

## 토큰화 모델 적용

```python
### 텍스트 문장 생성
text = '너와 함께한 시간 모두 눈부셨다. 날이 좋아서 날이 좋지 않아서 날이 적당해서 모든 날이 좋았다.'
```

```python
### 토큰화 모델 적용하기 (단어 사전의 크기 : 5000개)

# 단어 사전의 토큰 → 토큰화 + 단어 사전의 각 토큰별 인덱스를 참조 → 정수 인코딩
result1 = tokenizer1.encode(text)

# 결과 확인하기
print(result1)

print('-'*80)

# encoding의 결과 확인하기 → 속성 변수 : tokens
print(f'토큰화의 결과 : \n{result1.tokens}')
```

```python
### 토큰화 모델 적용하기 (단어 사전의 크기 : 20000개)

# 단어 사전의 토큰 → 토큰화 + 단어 사전의 각 토큰별 인덱스를 참조 → 정수 인코딩
result2 = tokenizer2.encode(text)

# 결과 확인하기
print(result2)

print('-'*80)

# encoding의 결과 확인하기 → 속성 변수 : tokens
print(f'토큰화의 결과 : \n{result2.tokens}')
```
