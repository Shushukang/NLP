```python
### 필요한 함수 임폴트
from transformers import pipeline
```

## text classification

```python
### pipeline 함수 호출, 분류 모델 생성(다운로드)

'''
# 조건 : 분류 모델, 토크나이저 지정(X)
'''

classifier = pipeline('text-classification')
```

```python
### test data 생성
test_data =["The restaurant is awesome","The restaurant is awful"]
```

```python
### 감성 분석 실행
# 원래라면 토크나이저로 토큰화 후 단어사전의 인덱스로 인코딩을 해서 임베딩을 진행해 Class 토큰이 생성되어 Dense layer로 처리가 되어야 함
result = classifier(test_data)

# 결과 확인하기
print(result)
```

```python
### 영어로 된 경제 기사의 감성 분석 전문 모델 사용

# pipeline 모델 지정 생성
model_name = 'ProsusAI/finbert'
finbert = pipeline(task = "text-classification", model = model_name)

# text 생성
text = "Stocks rallied and the British pound gained."

# 감성 분석 실행
result = finbert(text)

# 결과 확인하기
print(result)
```

```python
### 한글로 된 경제 기사의 감성 분석 전문 모델 사용

# pipeline 모델 지정 생성
model_name = 'snunlp/KR-FinBert-SC'
ko_finbert = pipeline(task = "text-classification", model = model_name)

# text 생성
text = "쇼핑몰 '알렛츠'도 갑자기 영업종료…'제2의 티메프 사태 우려'"

# 감성 분석 실행
result = ko_finbert(text)

# 결과 확인하기
print(result)
```

```python
### 한국어 감성 분석

#pipeline 모델 생성
model_name = 'doya/klue-sentiment-nsmc'
classifier1 = pipeline(task='text-classification', model=model_name)

# text 생성
text=['음악이 주가 된, 최고의 음악영화', '발연기 도저히 못보겠다 진짜 이렇게 연기를 못할거라곤 상상도 못했네']

# 감성 분석 실행
result1 = classifier1(text)

# 결과 확인하기
print(result1)
```

```python
### 한국어 감성 분석 → 파인 튜닝 모델 사용

# 필요한 함수 임폴트 (허깅페이스에 등록을 하지 않았으므로(비공개) 사용한 토크나이저와 학습 모델을 끌어와야 함)
from transformers import AutoTokenizer
from transformers import TFBertForSequenceClassification, TFElectraForSequenceClassification

# 토크나이저 생성
klue_tokenizer = AutoTokenizer.from_pretrained('klue/bert-base')
electra_tokenizer = AutoTokenizer.from_pretrained('beomi/KcELECTRA-base-v2022')

# bert의 한국어 분류 모델 생성
klue_bert_path = '/content/drive/MyDrive/딥러닝/tf_klue'
klue_bert = TFBertForSequenceClassification.from_pretrained(klue_bert_path)

# ELECTRA 한국어 분류 모델 생성
electra_bert_path = '/content/drive/MyDrive/딥러닝/tf_electra'
electra_bert = TFElectraForSequenceClassification.from_pretrained(electra_bert_path)

# pipeline 모델 생성
classifier2 = pipeline(task='text-classification', model=klue_bert, tokenizer=klue_tokenizer)
classifier3 = pipeline(task='text-classification', model=electra_bert, tokenizer=electra_tokenizer)

# text 생성
text=['음악이 주가 된, 최고의 음악영화', '발연기 도저히 못보겠다 진짜 이렇게 연기를 못할거라곤 상상도 못했네']

# 감성 분석 실행
result2 = classifier2(text)
result3 = classifier3(text)

# 결과 확인하기
print(result2)
print('-'*80)
print(result3)
```

## summarization

```python
### 기본 모델 생성
summarizer1 = pipeline(task='summarization')
```

```python
### 문서 요약에 사용할 텍스트 데이터 생성
text = '''The tower is 324 meters (1,063 ft) tall, about the same height ,
as an 81-storey building, and the tallest structure in Paris. Its base is square,
measuring 125 meters (410 ft) on each side. During its construction, the Eiffel
Tower surpassed the Washington Monument to become the tallest man-made structure
in the world, a title it held for 41 years until the Chrysler Building in New York
City was finished in 1930. It was the first structure to reach a height of 300 meters.
Due to the addition of a broadcasting aerial at the top of the tower in 1957, it is
now taller than the Chrysler Building by 5.2 meters (17 ft). Excluding transmitters,
the Eiffel Tower is the second tallest free-standing structure in France
after the Millau Viaduct.'''
```

```python
### 기본 모델을 이용한 문서 요약 실행
result1 = summarizer1(text)
print(f'기본 모델이 생성한 요약문 : \n{result1}')

print('-'*80)

# 결과값 → 텍스트 추출
result_text1 = result1[0]['summary_text']
print(f'기본 모델이 생성한 요약문 : \n{result_text1}')
```

```python
### 모델 지정 → 문서 요약 실행

# 모델 생성
model_name = 'facebook/bart-large-cnn'
summarizer2 = pipeline(task='summarization', model=model_name)

# 문서 요약 실행
result2 = summarizer2(text)

# 결과 확인하기
print(result2)

print('-'*80)

# 결과값 → 텍스트 추출
result_text2 = result2[0]['summary_text']
print(f'지정된 모델이 생성한 요약문 : \n{result_text2}')
```

## QuestionAnswering

```python
### pipeline 함수 호출, 모델 설정
qa = pipeline(task='question-answering')

# distilbert/distilbert-base-cased-distilled-squad - bert 모델 중 base(embedding vector size) 크기의 distilled(용량이 적게 개선된) squad(QA 모델)
```

```python
# 지문과 질문을 같이 입력하면 결과가 나옴
### QA에 사용할 텍스트 데이터 생성
question = "Which name is also used to describe the Amazon rainforest in English?"
context = """The Amazon rainforest (Portuguese: Floresta Amazônica or Amazônia; Spanish: Selva Amazónica, Amazonía or usually Amazonia; French: Forêt amazonienne;
Dutch:Amazoneregenwoud), also known in English as Amazonia or the Amazon Jungle, is a moist broadleaf forest that covers most of the Amazon basin of South America.
This basin encompasses 7,000,000 square kilometres (2,700,000 sq mi), of which 5,500,000 square kilometres (2,100,000 sq mi) are covered by the rainforest.
This region includes territory belonging to nine nations. The majority of the forest is contained within Brazil, with 60% of the rainforest,
followed by Peru with 13%, Colombia with 10%, and with minor amounts in Venezuela, Ecuador, Bolivia, Guyana, Suriname and French Guiana.
States or departments in four nations contain "Amazonas" in their names. The Amazon represents over half of the planet's remaining rainforests,
and comprises the largest and most biodiverse tract of tropical rainforest in the world, with an estimated 390 billion individual trees divided into 16,000 species."""
```

```python
### QA 실행 (질문 후 지문 순서로 입력 / 매개 변수와 함께 입력)
result = qa(question=question, context=context)

# 결과 확인하기 - 정답의 출력 형식은 딕셔너리
# print(result)
# 'score' : 정확도 - 기준을 알 수 없음
# 'start'/'end' : 단어의 시작 인덱스 / 단어의 끝 인덱스(미포함)
# 'answer' : 필요한 정답

# 결과 확인하기 → 'answer' 추출
print(result['answer'])
```

```python
### qa 모델 지정, pipeline 호출, 모델 생성
model_name = 'deepset/roberta-base-squad2'
qa1 = pipeline(task='question-answering', model=model_name)
```

```python
### QA 실행
result1 = qa1(question=question, context=context)

# 결과 확인하기 → 'answer' 추출
print(result1['answer'])
```

## text generation

```python
### pipeline 함수 호출, 모델 생성
generator = pipeline(task='text-generation')
```
