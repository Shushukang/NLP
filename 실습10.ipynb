## 한글 문장 유사도 측정

```python
### 사전 학습된 한국어 SentenceTransformer 모델 생성
# 'base' = 임베딩 벡터의 사이즈가 768 자릿임
model_name = 'ddobokki/klue-roberta-base-nli-sts'
model_kor = SentenceTransformer(model_name)
```

```python
### 한글 텍스트 데이터 생성
ko_sentences = ["직원이 무단 퇴사를 했는데 손해 배상 청구할 수 있나요?",
                "무단 퇴사한 직원에 대한 손해 배상 청구가 가능한가요?"]
```

```python
### 한글 텍스트 데이터 생성(서로 무관한 데이터)
ko_sentences1 = ["직원이 무단 퇴사를 했는데 손해 배상 청구할 수 있나요?",
                "오늘도 날씨가 매우 덥고 소나기가 내리는 등 동남아 날씨 같아요."]
```

```python
### 문장 임베딩 벡터 생성
ko_embeddings = model_kor.encode(ko_sentences)

# 결과 확인하기
print(f'한글 텍스트의 임베딩 행렬의 모양 : {ko_embeddings.shape}')
print('-'*80)
print(f'한글 텍스트의 첫 번째 문장의 임베딩 벡터 : \n{ko_embeddings[0, :]}')
print('-'*80)
print(f'한글 텍스트의 두 번째 문장의 임베딩 벡터 : \n{ko_embeddings[1, :]}')
```

```python
### 문장 임베딩 벡터 생성
ko_embeddings1 = model_kor.encode(ko_sentences1)

# 결과 확인하기
print(f'한글 텍스트의 임베딩 행렬의 모양 : {ko_embeddings1.shape}')
print('-'*80)
print(f'한글 텍스트의 첫 번째 문장의 임베딩 벡터 : \n{ko_embeddings1[0, :]}')
print('-'*80)
print(f'한글 텍스트의 두 번째 문장의 임베딩 벡터 : \n{ko_embeddings1[1, :]}')
```

```python
### 코사인 유사도 측정

# 필요한 라이브러리 임폴트
from sklearn.metrics.pairwise import cosine_similarity

# 1차원 배열 → 2차원 배열로 전환
embedding1 = ko_embeddings[0,:].reshape((1,768))
embedding2 = ko_embeddings[1,:].reshape((1,768))

# cosine_similarity(x, y) 호출 : x,y는 반드시 2차원 배열이어야 함
similarity = cosine_similarity(embedding1, embedding2)

# 결과 확인하기
print(similarity)
```

```python
### 유사하지 않은 두 문장에 대한 코사인 유사도 측정

# 1차원 배열 → 2차원 배열로 전환
embedding3 = ko_embeddings1[0,:].reshape((1,768))
embedding4 = ko_embeddings1[1,:].reshape((1,768))

# cosine_similarity(x, y) 호출 : x,y는 반드시 2차원 배열이어야 함
similarity1 = cosine_similarity(embedding3, embedding4)

# 결과 확인하기
print(similarity1)
```

## 영어 문장 유사도 측정

```python
### 사전 학습된 영어 SentenceTransformer 모델 생성
model_name='all-MiniLM-L12-v2'
model_eng = SentenceTransformer(model_name)
```

```python
### 영문 텍스트 데이터 생성
en_sentences = ["What should I do to be a great scientist?",
                "How can I be a good scientist?"]
```

```python
### 문장 임베딩 행렬 생성
en_embeddings = model_eng.encode(en_sentences)

# 결과 확인하기
print(f'영문 텍스트의 임베딩 행렬의 모양 : {en_embeddings.shape}')
print('-'*80)
print(f'영문 텍스트의 첫 번째 문장의 임베딩 벡터 : \n{en_embeddings[0, :]}')
print('-'*80)
print(f'영문 텍스트의 두 번째 문장의 임베딩 벡터 : \n{en_embeddings[1, :]}')
```

```python
### 코사인 유사도 측정하기

# 1차원 배열 → 2차원 배열로 변환
embedding1 = en_embeddings[0,:].reshape((1,384))
embedding2 = en_embeddings[1,:].reshape((1,384))

# cosine_similarity(X, Y) 호출 : x,y는 반드시 2차원 배열이어야 함
similarity = cosine_similarity(embedding1, embedding2)
print(similarity)
```

# 문장 유사도를 이용한 추천 시스템

```python
### 데이터 불러오기

# 필요한 라이브러리 임폴트
import pandas as pd

# 파일 경로 설정
file_path = '/content/drive/MyDrive/딥러닝/movies_metadata.csv'

# pd.read_csv() 함수 호출 → DataFrame 생성
df = pd.read_csv(file_path)

# 결과 확인하기
print(df)
```

```python
### 영화 소개(overview), 제목(title) 컬럼 추출 → fancy indexing
df1 = df.loc[:,['overview','title']]
print(df1)
```

```python
### 누락 데이터 제거

# dropna() 함수 호출
# print(df1.isnull().sum())
df2 = df1.dropna(ignore_index=True)

# 결과 확인하기
print(df2)
```

```python
### data의 수를 10000개로 축소
df_data = df2.iloc[0:10000, :]

# 결과 확인하기
print(df_data)
```

```python
### 사전 학습된 SentenceTransformer 모델 생성
model_name='all-MiniLM-L12-v2'
model = SentenceTransformer(model_name)
```

```python
### 문장 임베딩 행렬 생성

# overview 컬럼 → 배열 생성
input = df_data.loc[:, 'overview'].values
# print(f'입력 데이터 : \n{input[:10]}')

# model.encode() 호출
embeddings = model.encode(input)

# 결과 확인하기
print(f'생성된 임베딩 행렬의 모양 : {embeddings.shape}')
print('-'*80)
print(f'첫 번째 영화 소개글에 대한 임베딩 벡터 : \n{embeddings[0]}')
```

```python
### 코사인 유사도 측정 → 10000개의 영화 소개글에 대한 측정

# 필요한 함수 임폴트
from sklearn.metrics.pairwise import cosine_similarity

# cosine_similarity(X,Y) 함수 호출
similarity = cosine_similarity(embeddings, embeddings)

# 결과 확인하기
print(f'유사도 측정의 결과 : \n{similarity}')
print('-'*80)
print(f'유사도 측정의 결과 값의 모양 : \n{similarity.shape}')
```

```python
### 코사인 유사도 → DataFrame 생성

# index, columns 설정
index = df_data.loc[:, 'title'].values
columns = df_data.loc[:, 'title'].values

# pd.DataFrame() 함수 호출
df_sim = pd.DataFrame(data=similarity, index = index, columns = columns)

# 결과 확인하기
# print(df_sim)
df_sim
```

```python
### 특정 영화 선택 → 특정 영화와 나머지 영화 간의 코사인 유사도를 내림 차순으로 정렬

# 특정 영화 선택 → 특정 컬럼 인덱싱 → 코사인 유사도 top10 추출
# 음수가 존재하므로 정반대의 영화도 추출 가능(ascending=True)
print(df_sim.loc[:, 'Toy Story'].sort_values(ascending=False).iloc[1:11])
print('-'*80)
print(df_sim.loc[:, 'Toy Story'].sort_values(ascending=True).iloc[0:10])
```

```python
### 특정 영화 기준 → 줄거리가 유사한 영화 추천 함수 생성

def recommend(title):
  top10 = df_sim.loc[:, title].sort_values(ascending=False).iloc[1:11]
  return top10
```

```python
### 줄거리가 유사한 영화 추천 함수 실행
name = 'GoldenEye'
top10 = recommend(title = name)
print(f'결과 확인하기 : \n{top10}')
```
